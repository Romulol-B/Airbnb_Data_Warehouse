```python
import pandas as pd
import os

# 1. Defina o caminho para a pasta com os arquivos .txt
#    O '.' significa 'a pasta atual onde o script está rodando'.
#    'meus_txts' é o nome da subpasta que contém os arquivos.
caminho_da_pasta = f'{local_path}data/external' 

# 2. Defina o nome do arquivo CSV final que será gerado
arquivo_csv_final = f'{local_path}/data/raw/dados_pluviometricos_consolidados.csv'
import pandas as pd
import os
import re # <-- Adicionado: Importa a biblioteca de expressões regulares

# 3. Crie uma lista vazia para armazenar os dados de cada arquivo
lista_de_dados = []

print(f"Iniciando processo na pasta: '{caminho_da_pasta}'")

if not os.path.isdir(caminho_da_pasta):
    print(f"Erro: A pasta '{caminho_da_pasta}' não foi encontrada.")
    print("Por favor, crie a pasta e coloque seus arquivos .txt dentro dela.")
else:
    for nome_do_arquivo in os.listdir(caminho_da_pasta):
        if nome_do_arquivo.endswith('Plv.txt'):
            #print(f"Processando arquivo: {nome_do_arquivo}...")
            
            caminho_completo_arquivo = os.path.join(caminho_da_pasta, nome_do_arquivo)
            
            try:
                # --- LÓGICA DE EXTRAÇÃO DO NOME MODIFICADA ---
                
                # Pega o nome do arquivo sem a extensão .txt
                nome_base = os.path.splitext(nome_do_arquivo)[0]
                
                # Divide o nome_base no primeiro dígito que encontrar.
                # Ex: 'abc_def_20150101_plv' -> ['abc_def_', '0150101_plv']
                partes_nome = re.split(r'[0-9]', nome_base, maxsplit=1)
                
                # A primeira parte da divisão é o que queremos
                localizacao_bruta = partes_nome[0]
                
                # Remove qualquer underline que possa ter ficado no final
                # Ex: 'abc_def_' -> 'abc_def'
                localizacao = localizacao_bruta.rstrip('_')

                # --- FIM DA MODIFICAÇÃO ---

                # Lê o arquivo txt usando a expressão regular para o separador
                dados_arquivo = pd.read_csv(
                    caminho_completo_arquivo, 
                    skiprows=3, 
                    sep=r'\s{2,}',
                    engine='python'
                )
                
                # Adiciona a nova coluna 'localizacao' com o nome limpo
                dados_arquivo = dados_arquivo.assign(localizacao=localizacao)
                colunas = list(dados_arquivo.columns)
                dados_arquivo = dados_arquivo[[colunas[-1]] + colunas[:-1]] 

                lista_de_dados.append(dados_arquivo)

            except Exception as e:
                print(f"  -> Erro ao processar o arquivo {nome_do_arquivo}: {e}")

    if lista_de_dados:
        dados_consolidados = pd.concat(lista_de_dados, ignore_index=True)
        dados_consolidados.to_csv(arquivo_csv_final, index=False)
    else:
        print("\nNenhum arquivo .txt foi encontrado ou processado na pasta.")
```
